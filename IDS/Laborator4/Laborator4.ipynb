1. Incarcati setul de date in format tsv [https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) intr-un obiect Pandas DataFrame.

```python
import pandas as pd

data_set = pd.read_csv('chipotle.tsv', sep='\t', header=0)
type(data_set)
```

    A. Afisati primele 5 linii din tabel


```python
data_set.head()
```

    B. Care este numarul de date din DataFrame?


```python
""" count() method doesn't count NaN values, that's why choice_description has fewer values than the others. 
Using len() method we know for sure the number of rows in data_set """
print(len(data_set))
print(data_set.count())
```

    C. Comparati iesirea metodelor `info()` si `describe()`.


```python
""" info() method prints information about data_set like index and column dtypes, non-null values and memory usage. 
On the other hand, describe() method generates some basic statistical details like percentile, mean, maximum and minimum values
about data_set. Also, the last method seems to be applied only to columns with numeric values. """

data_set.info()
data_set.describe()
```

    D. Care e numarul de coloane? afisati numele lor.


```python
print(f"Number of columns = {data_set.shape[1]}")
data_set.columns.values
```

    E. Cum e indexat setul de date? sunt valorile din index unice?


```python
""" This data_set doesn't contain unique values as index """
print(type(data_set.order_id))
data_set.order_id
```

    F. Care sunt marfurile vandute, in ordine alfabetica? Cate sunt?


```python
print(data_set.sort_values('item_name').item_name)
print(f"There are {len(data_set.item_name)} items")
```

    G. Cate obiecte (items) au fost comandate, in total - suma pe coloana 'quantity'?


```python
data_set.quantity.sum()
```

H. Transformati intr-un ant dataframe coloana 'item_price' din stringuri in valori floating point. Care sunt preturile (minim, maxim, mediu, median) cu care s-a vandut produsul 'Chicken Crispy Tacos'? Cate astfel de produse au fost comandate in total? Care sunt veniturile obtinute din vanzarea acestui produs? *Nota*: puteti folosi metoda `describe()` a unui dataframe/series.

```python
price = pd.DataFrame(data_set.item_price.str.replace('$', '').astype('float'))
print(f'Total price = {price[data_set.item_name == "Chicken Crispy Tacos"].sum()}')
price[(data_set.item_name == 'Chicken Crispy Tacos')].describe()
# print(price[(data_set.item_name == 'Chicken Crispy Tacos')])
```

I. Care sunt cele mai comandate 3 marfuri? Folositi `groupby()` pentru grupare, ne intereseaza cantitatile cele mai mari. *Nota:* sectiunea 'Aggregation and Grouping' din ref bibliografica 1 din curs e utila pentru intelegerea agregarilor.

```python
data_set.groupby('item_name').aggregate(sum)['quantity'].sort_values(inplace = False, ascending = False)[:3]
```

    J. De cate ori s-a comandat 'Veggie Salad Bowl'?

```python
data_set[data_set.item_name == 'Veggie Salad Bowl'].quantity.sum()
```

    K. De cate ori intr-o comanda s-au cerut mai mult de o 'Canned Soda'?


```python
len(data_set[(data_set.item_name == 'Canned Soda') & (data_set.quantity > 1)])
```

L. Care sunt banii incasati pentru fiecare comanda? atentie, pentru fiecare item comandat trebuie luat in considerare si cantitatea. Puteti crea coloane auxiliare (pret transformat din string in float, pret total pe acel item = pret_unitar * cantitate etc.)

```python
price = data_set.item_price.str.replace('$', '').astype('float')
data_set['items_price'] = price.multiply(data_set.quantity)
data_set.groupby('order_id')['items_price'].sum()
```

2. Incarcati setul de date de la https://raw.githubusercontent.com/jokecamp/FootballData/master/UEFA_European_Championship/Euro%202012/Euro%202012%20stats%20TEAM.csv intr-un obiect Pandas DataFrame

```python
import pandas as pd

data_set = pd.read_csv('Euro 2012 stats TEAM.csv', header=0)
data_set
```

    A. Cate echipe au participat la Euro 2012?


```python
len(data_set.index)
```

    B. Sa se sorteze echipele dupa numarul de cartonase rosii, apoi dupa cele galbene, descrescator - sortare dupa criteriu dublu.

```python
data_set.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

    C. Care e numarul mediu de cartonase rosii date? Dar mediana?


```python
print('Mean = {} and Median = {}'.format(data_set["Red Cards"].mean(), data_set["Red Cards"].median()))
```

    D. Care sunt echipele care au luat mai multe cartonase rosii decat media?


```python
mean = data_set["Red Cards"].mean()
data_set[data_set['Red Cards'] > mean].Team
```

    F. Care sunt Shooting accuracy pentru Anglia, Grecia, Rusia?

```python
teams = ['England', 'Greece', 'Russia']
data_set.loc[data_set.Team.isin(teams),['Team','Shooting Accuracy']]
```

# Exercitiul 3.

3. Incarcati setul de date de la [https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). Folosind grupari (`groupby`) dupa o documentare serioasa in sectiunea 'Aggregation and Grouping' din ref bibliografica [1] din primul curs, rezolvati:

```python
import pandas as pd

data_set = pd.read_csv('drinks.csv', header=0)
data_set
```

    A. Pe care continent se bea mai multa bere, in medie?


```python
data_set.groupby('continent')['beer_servings'].aggregate(sum).sort_values(ascending=False)[:1]/len(data_set['continent'])
```

    B. Pentru fiecare continent sa se afiseze statisticile pentru consumul de vin.


```python
data_set.groupby('continent')['wine_servings'].describe()
```

    C. Sa se tipareasca min, max, media pentru consumul de bere pe fiecare continent. Folositi functia `agg`.

```python
import numpy as np

data_set.groupby('continent')['beer_servings'].agg([min,max,np.mean])
```
