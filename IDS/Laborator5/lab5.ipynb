Incarcati un set de date numerice, multivariat dintr-unul din depozitele de date mentionate in cursul 5. Folosind metodele prezentate la Visualization, Pandas Dataframe: Plot Examples with Matplotlib and Pyplot, reprezentati valorile pe coloane. Comentati (in text) cele observate pe baza reprezentarilor (de exemplu: exista valori outlier, valorile sunt concentrate intr-un anumit interval, vlorile sunt grupate in catve locatii etc).

Attribute Information:

    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per 10,000 dolars
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in 1000 dolars


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
address = 'Housing.csv'

dataframe = pd.read_csv(address, sep=',')
dataframe
```

```python
def normalizare_date(x):
    minimum = x.min(axis=0)
    maximum = x.max(axis=0)
    modification = minimum != maximum
    x[:,modification] = (x[:,modification] - minimum[modification]) / (maximum[modification] - minimum[modification])
    return x, minimum, maximum
```

```python
uniform_values, minim, maxim = normalizare_date(dataframe.values)
```

```python
uniform_dataframe = pd.DataFrame({
    'CRIM':uniform_values[:, 0],
    'ZN':uniform_values[:, 1],
    'INDUS':uniform_values[:, 2],
    'CHAS':uniform_values[:, 3],
    'NOX':uniform_values[:, 4],
    'RM':uniform_values[:, 5],
    'AGE':uniform_values[:, 6],
    'DIS':uniform_values[:, 7],
    'RAD':uniform_values[:, 8],
    'TAX':uniform_values[:, 9],
    'PTRATIO':uniform_values[:, 10],
    'B':uniform_values[:, 11],
    'LSTAT':uniform_values[:, 12],
    'MEDV':uniform_values[:, 13]
})
```

```python
'''
Graficul contine toate valorile normalizate a 
dataframe-ului. Dupa cum se poate prea bine observa 
nu toate datele sunt in concordanta cu celelalte si
o sa trebuiasca sa gasim unele atribute separate care
par ca au sens cu altele.
'''

with pd.plotting.plot_params.use('x_compat', True):
    uniform_dataframe.CRIM.plot(color='#191970', figsize=(20, 20))
    uniform_dataframe.ZN.plot(color='#000080')
    uniform_dataframe.INDUS.plot(color='#6495ed')
    uniform_dataframe.CHAS.plot(color='#483d8b')
    uniform_dataframe.NOX.plot(color='#6a5acd')
    uniform_dataframe.RM.plot(color='#7b68ee')
    uniform_dataframe.AGE.plot(color='#8470ff')
    uniform_dataframe.DIS.plot(color='#0000cd')
    uniform_dataframe.RAD.plot(color='#4169e1')
    uniform_dataframe.TAX.plot(color='#0000ff')
    uniform_dataframe.PTRATIO.plot(color='#1e90ff')
    uniform_dataframe.B.plot(color='#00bfff')
    uniform_dataframe.LSTAT.plot(color='#87ceeb')
    uniform_dataframe.MEDV.plot(color='#00ffff')
```

```python
dataframe.plot.area(stacked=False, figsize=(20, 20))
```

```python
'''
Aici sunt afisate toate scateruruile formate din fiecare atribut si obesrvam
ca exista multe legaturi intre atributele din dataframe. Cum ar fi AGE cu LSTAT
sau MEDV cu LSTAT
'''
pd.plotting.scatter_matrix(uniform_dataframe, alpha=0.2, figsize=(20, 20), diagonal='kde')
plt.show()
```

```python
'''
Acum ca avem afisat in grafice separate valorile dataframe-ului
putem sa distingem unele atribute care par sa aibe valori asemanatoare,
cum ar fi RAD, TAX, PTRATIO si INDUS, in timp ce
CRIM nu pare sa aibe legatura cu vreun alt atribut.
'''
uniform_dataframe.plot(subplots=True, figsize=(20, 20))
plt.show()
```
